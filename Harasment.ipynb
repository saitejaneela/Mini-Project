```python
import matplotlib.pyplot as plt
import pandas as pd
dataset=pd.read_csv('Andhars.csv')
left=dataset.iloc[:,0]
height=dataset.iloc[:,1]
tick_label=dataset.iloc[:,2]
dataset.iloc[0,:]
plt.bar(left, height, tick_label = tick_label,width = 0.8, color = ['red', 'green'])
plt.xlabel('Years')
plt.ylabel('Harrassment')
plt.savefig("Andhars.png")
plt.title('Andhrapradesh Harrassment')
plt.show()
```
